## Functions to load images and labels in a NN Format

```python
def get_data(path,n):
    f = gzip.open(path,'r')
    image_size = 250
    num_images = n
    f.read(16)
    buf = f.read(image_size * image_size * num_images)
    data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
    data = data.reshape(num_images, image_size, image_size)
    return data
```

```python
def get_labels(path,n):
    to_return = []
    f = gzip.open(path,'r')
    f.read(8)
    for i in range(0,n):   
        buf = f.read(1)
        label = np.frombuffer(buf, dtype=np.uint8).astype(np.int64)
        to_return.append(label)
    return np.squeeze(np.array(to_return)) 
```

```python
# Defining paths and number of images
# ---------------------- TRAIN --------------------------
train_images_path = './data/converted-keras/train-images-idx3-ubyte.gz'
train_qty = 723*3;
train_labels_path = './data/converted-keras/train-labels-idx1-ubyte.gz'
# ---------------------- TEST --------------------------
test_images_path = './data/converted-keras/test-images-idx3-ubyte.gz'
test_qty = 320*3;
test_labels_path = './data/converted-keras/test-labels-idx1-ubyte.gz'
```

```python
# defining variables to train & test model
train_images = get_data(train_images_path, train_qty);
train_labels = get_labels(train_labels_path, train_qty);
```

```python
test_images = get_data(test_images_path, test_qty);
test_labels = get_labels(test_labels_path, test_qty);
```

## Creating Model with Tensor Flow & Keras

```python
# creating the model
model = keras.Sequential([
    # inserting refformated data as a Flat array
    keras.layers.Flatten(input_shape=(250,250)),
    # 128 nodes or neurons
    keras.layers.Dense(182, activation=tf.nn.relu),
    # 3 nodes (neurons) which represent the final probability output.
    keras.layers.Dense(3, activation=tf.nn.softmax)
])
```

### Setting up parameters

```python
model.compile(
    # model optimizer
    optimizer='adam',
    
    # loss function or how will we measure accuracy in order to adjust.
    loss='sparse_categorical_crossentropy',
    #loss='mean_squared_error',
    #loss='categorical_crossentropy',
    
    # how to evaluate
    metrics=['accuracy']   
)
```

### Train

```python
# train the model
model.fit(train_images, train_labels, epochs=20)
```

### Testing Model

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test Loss: ',test_loss,' Test Accuracy: ',test_acc)
```

## Plot image data against prediction to check accuracy

```python
def plot_image(i,predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    plt.imshow(img, cmap=plt.cm.binary)
    
    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'green'
    else:
        color = 'red'
    
    plt.xlabel("{} {:2.0f}% ({})".format(
            class_names[predicted_label],
            100*np.max(predictions_array),
        class_names[true_label]), color=color)
```

```python
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#FF0000")
    plt.ylim([0,1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('green')
```

```python
rows = 5
cols = 3
num_images = rows * cols
plt.figure(figsize = (3*2*cols, 3*rows) )
for i in range(num_images):
    plt.subplot(rows,2*cols,2*i+1)
    plot_image(i, predictions, test_labels, test_images)

plt.show()
```
